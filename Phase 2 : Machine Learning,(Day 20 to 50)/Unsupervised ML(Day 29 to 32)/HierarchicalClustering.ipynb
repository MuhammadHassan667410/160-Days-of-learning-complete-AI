```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv(r"C:\Users\taj\Desktop\VSCodesFOLDER\Mall_Customers.csv")
df.head()
```

```python
df.info()
```

```python
df1 = df.drop(['CustomerID', 'Genre', 'Age'], axis=1)
df1.head()
```

```python
import scipy.cluster.hierarchy as shc
```

```python
plt.figure(figsize=(10,7))
plt.title('Customer Dendogram')
dend = shc.dendrogram(shc.linkage(df1, method='ward'))
```

```python
# identify if we can get more than three clusters
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='ward')
labels_ = model.fit_predict(df1)
```

```python
labels_
```

```python
import seaborn as sns

sns.scatterplot(x=df1.iloc[:, 0], y=df1.iloc[:, 1], hue=model.labels_, palette='Set2')
plt.title("Agglomerative Clustering")

```
