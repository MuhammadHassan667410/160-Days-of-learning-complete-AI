```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/housing.csv")
```

```python
df
```

```python
lat_long = df[['longitude', 'latitude']]
long, lat = df.longitude, df.latitude
plt.scatter(long, lat)
```

```python
from sklearn.cluster import DBSCAN
```

```python
X = lat_long.to_numpy()
X.shape
```

```python
model = DBSCAN(eps=0.2, min_samples=15)
model.fit(X)
```

```python
df['clusters'] = model.labels_
```

```python
df
```

```python
df.drop(df[['ocean_proximity']], axis=1, inplace=True)
df
```

```python
df['clusters'].value_counts()
```

```python
import plotly.express as px

fig = px.scatter(
    df,
    x='longitude',
    y='latitude',
    color='clusters'
)
fig.show()
```

```python
from sklearn.metrics import silhouette_score as ss

ss(X, df['clusters'])
```

```python
epsilons = np.linspace(0.01, 1, num=15)
epsilons
```

```python
min_samples = np.arange(2, 20, step=3)
min_samples
```

```python
import itertools

combinations = list(itertools.product(epsilons, min_samples))
combinations
```

```python
N = len(combinations)
N
```

```python
def get_scores_and_labels(combinations, X):
    scores = []
    all_labels_list = []

    for i, (eps, num_samples) in enumerate(combinations):
        db_model = DBSCAN(eps=eps, min_samples=num_samples).fit(X)
        labels = db_model.labels_
        labels_set = set(labels)
        num_clusters = len(labels_set)
        if -1 in labels_set:
            num_clusters -= 1
        if (num_clusters < 2) or (num_clusters > 50):
            scores.append(-10)
            all_labels_list.append('bad')
        else:
            scores.append(ss(X, labels))
            all_labels_list.append(labels)


    best_index = np.argmax(scores)
    best_parameters = combinations[best_index]
    best_labels = all_labels_list[best_index]
    best_score = scores[best_index]

    return {'Best Epsilons': best_parameters[0],
            'Best min_samples': best_parameters[1],
            'Best Labels': best_labels,
            'Best Score': best_score}

best_dict = get_scores_and_labels(combinations, X)
```

```python
best_dict
```

```python
df['cluster'] = best_dict['Best Labels']
df['cluster'].value_counts()
```

```python
fig = px.scatter(
    df,
    x='longitude',
    y='latitude',
    color='cluster'
)
fig.show()
```

```python
df.drop(df[['clusters']], axis=1, inplace=True)
df
```
