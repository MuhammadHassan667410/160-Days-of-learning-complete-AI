```python
from sklearn.cluster import KMeans
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
%matplotlib inline
```

```python
data = pd.read_csv('income.csv')
```

```python
data
```

```python
plt.scatter(data['Age'], data['Income($)'])
```

```python
km = KMeans(n_clusters = 3)
km
```

```python
y_pred = km.fit_predict(data[['Age','Income($)']])
```

```python
y_pred
```

```python
data['Cluster'] = y_pred
data
```

```python
d1 = data[data.Cluster==0]
d2 = data[data.Cluster==1]
d3 = data[data.Cluster==2]
```

```python
plt.scatter(d1.Age, d1['Income($)'], label='Cluster 1')
plt.scatter(d2.Age, d2['Income($)'], label='Cluster 2')
plt.scatter(d3.Age, d3['Income($)'], label='Cluster 3')
plt.xlabel('Age')
plt.ylabel('Income in $')
plt.legend()
plt.show()
```

```python
scaler = MinMaxScaler()
data['Income($)'] = scaler.fit_transform(data[['Income($)']])
data['Age'] = scaler.fit_transform(data[['Age']])
```

```python
data
```

```python
km = KMeans(n_clusters= 3)
y_pred = km.fit_predict(data[['Age', 'Income($)']])
y_pred
```

```python
data.Cluster = y_pred
data
```

```python
km.cluster_centers_
```

```python
plt.scatter(d1.Age, d1['Income($)'], label='Cluster 1')
plt.scatter(d2.Age, d2['Income($)'], label='Cluster 2')
plt.scatter(d3.Age, d3['Income($)'], label='Cluster 3')
plt.scatter(km.cluster_centers_ [:,0], km.cluster_centers_ [:,1], marker='*', label='Center')
plt.xlabel('Age')
plt.ylabel('Income in $')
plt.legend()
plt.show()
```

```python
k_range = range(1,10)
sse = []
for k in k_range:
    km = KMeans(n_clusters = k)
    km.fit(data[['Age', 'Income($)']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('Sum of Squared Errors')
plt.plot(k_range, sse, c = 'k', marker = 'o')
```
